```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data = pd.read_csv("./datasets/house_prices.csv")
data
```

```python
plt.xlabel('Area of living room (sqft)')
plt.ylabel('Price (USD)')
plt.scatter(data.sqft_living, data.price, color='red', marker='o')
```

```python
reg = linear_model.LinearRegression()

# train the model (runs gradient descent behind the scenes)
reg.fit(data[['sqft_living']], data.price)
```

```python
output = reg.predict([[8000]])
output
```

```python
# gives value of parameters (theta)
print(reg.coef_)

# gives the intercept
print(reg.intercept_)
```

```python
# plotting simple linear regression
plt.xlabel('Area of living room (sqft)')
plt.ylabel('Price (USD)')
plt.scatter(data.sqft_living, data.price, color='red', marker='o')
plt.plot(data.sqft_living, reg.predict(data[['sqft_living']]), color='blue')
```
