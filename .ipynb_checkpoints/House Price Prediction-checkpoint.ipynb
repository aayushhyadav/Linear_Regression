```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn import model_selection
from sklearn import metrics
```

```python
data = pd.read_csv("./datasets/house_prices.csv")
data
```

```python
features = ['sqft_living']
X = data[features]
y = data['price']

# spliting the data into training and testing sets
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y)
```

```python
plt.xlabel('Area of living room (sqft)')
plt.ylabel('Price (USD)')
plt.scatter(data.sqft_living, data.price, color='red', marker='o')
```

```python
reg = linear_model.LinearRegression()

# train the model (runs gradient descent behind the scenes)
reg.fit(X_train, y_train)
```

```python
output = reg.predict([[1340]])
output
```

```python
# gives value of parameters (theta)
print(reg.coef_)

# gives the intercept
print(reg.intercept_)
```

```python
y_prediction = reg.predict(X_test)
```

```python
# plotting simple linear regression
plt.xlabel('Area of living room (sqft)')
plt.ylabel('Price (USD)')
plt.scatter(data.sqft_living, data.price, color='red', marker='o')
plt.plot(X_train, reg.predict(X_train), color='blue')
```

```python
reg_error = np.sqrt(metrics.mean_squared_error(y_test, y_prediction))
reg_error
```

```python
# checking for null values
data.isnull().sum()
```

```python
# adding two new features age and renovation age
data_size = len(data)
entry = 0
data['age'] = np.zeros(data_size)
data['renovation_age'] = np.zeros(data_size)

while entry < data_size:
    data.loc[entry, 'age'] = 2022 - data.loc[entry, 'yr_built']
    
    if data.loc[entry, 'yr_renovated'] != 0:
        data.loc[entry, 'renovation_age'] = 2022 - data.loc[entry, 'yr_renovated']
    else:
        data.loc[entry, 'renovation_age'] = data.loc[entry, 'age']

    entry += 1

data
```

```python
# analysing relationship between price and different features
sns.pairplot(data, x_vars = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'sqft_above', 'sqft_basement', 'age', 'renovation_age'], y_vars = 'price', kind = 'reg')
```

```python
multi_features = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'sqft_above', 'sqft_basement', 'age', 'renovation_age']
multi_X = data[multi_features]
multi_y = data['price']
multi_X_train, multi_X_test, multi_y_train, multi_y_test = model_selection.train_test_split(multi_X, multi_y)
```

```python
multi_reg = linear_model.LinearRegression()

# train the model using multiple features
multi_reg.fit(multi_X_train, multi_y_train)
```

```python
multi_y_prediction = multi_reg.predict(multi_X_test)
```

```python
print(multi_reg.coef_)
print(multi_reg.intercept_)
```

```python
multi_reg_error = np.sqrt(metrics.mean_squared_error(multi_y_prediction, multi_y_test))
multi_reg_error
```
